```python
wheat_genomes_accessions = []

for genome in wheat_genomes:
    wheat_genomes_accessions.append('GCA_'+genome.split('_')[1])
```

```python
accessions_not_found = []
accessions_found = {}


for accession in wheat_genomes_accessions:
    accessions_found[accession] = {}
    print(accession)
    ## search for SRA to get ID number
    handle = Entrez.esearch(db="assembly", term=accession)
    summary = Entrez.read(handle)
    ## then get all info associated with ID number
    esummary = Entrez.esummary(db="assembly", id=summary['IdList'][0], retmode='xml')
    record = Entrez.read(esummary)
    # grab bioproject ID and then look that up
    bioproject_id = record['DocumentSummarySet']['DocumentSummary'][0]['BioSampleId']
    esummary = Entrez.esummary(db="biosample", id=bioproject_id, retmode='xml')
    record = Entrez.read(esummary)
    # parse broken xml file
    string = record['DocumentSummarySet']['DocumentSummary'][0]['SampleData']
    result = re.search('<Id db="SRA">(.*)</Id>', string)
    if result:
        sra_accession = result.group(1)
        handle = Entrez.esearch(db="sra", term=sra_accession)
        summary = Entrez.read(handle)
        ## then get all info associated with ID number
        ids = summary['IdList']
        for id in ids:
            esummary = Entrez.esummary(db="sra", id=id, retmode='xml')
            record = Entrez.read(esummary)
            # run id
            string = record[0]['Runs']
            result = re.search('<Run acc="(.*)" total_spots="', string)
            run_id = result.group(1)
            print(run_id)
            # sequencing tech
            string = record[0]['ExpXml']
            result = re.search('<Platform instrument_model=(.*)</Platform>', string)
            platform = result.group(1)
            print(platform)
            # number of bases
            string = record[0]['ExpXml']
            result = re.search('total_bases="(.*)" total_size=', string)
            bases = int(result.group(1))
            print(bases)
            if bases != 0: ## some records are empty unfortunately...
                accessions_found[accession][run_id] = platform
    else:
        print('not found')
        del accessions_found[accession]
        accessions_not_found.append(accession)
```

```python
runs_to_download = []

for accession in accessions_found:
    if len(accessions_found[accession]) == 1:
        for key in accessions_found[accession]:
            if "ILLUMINA" in accessions_found[accession][key]:
                runs_to_download.append(key)
    else:
        illumina_runs = []
        for key in accessions_found[accession]:
            if "ILLUMINA" in accessions_found[accession][key]:
                illumina_runs.append(key)
        if len(illumina_runs) == 1:
            runs_to_download.append(illumina_runs[0])
        else:
            ## deal with some exceptions, three of them are mini chrom sequencing so drop those
            ## one is worth keeping though
            if accession == 'GCA_004785725.2':
                runs_to_download.append('SRR6232156')
```
