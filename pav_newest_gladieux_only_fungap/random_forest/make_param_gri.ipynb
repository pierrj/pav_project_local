```python
import numpy as np
from itertools import product
from sklearn.model_selection import ParameterGrid
```

```python
approach = ["RF", "SMOTE"]
majority_fraction = [0.05, 0.25, 0.5, 1.0]
```

```python
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 4000, num = 10)]
```

```python
min_samples_split = [2, 5, 10]
```

```python
min_samples_leaf = [1, 2, 4]
```

```python
max_features = ["sqrt", "log2", None]
```

```python
max_depth = [int(x) for x in np.linspace(10, 200, num = 9)]
max_depth.append(None)
```

```python
bootstrap = [True,False]
```

```python
print(len(list(product(approach,
            majority_fraction,
            n_estimators,
            min_samples_split,
            min_samples_leaf,
            max_features,
            max_depth,
            bootstrap))))
```

```python
grid = product(approach,
            majority_fraction,
            n_estimators,
            min_samples_split,
            min_samples_leaf,
            max_features,
            max_depth,
            bootstrap)
```
